## Association Rules

```python
apriori_df = apriori(new_df,
                     min_support=0.001,
                     use_colnames=True,
                     low_memory=True)
rules = association_rules(apriori_df,
                          metric='support',
                          min_threshold=0.001)
```

We rank invoices based on lift metric.

```python
sorted_rules = rules.sort_values('lift', ascending=False)
sorted_rules
```

```python

recommendation_list = []
for i, product in enumerate(sorted_rules["antecedents"]):
    for j in list(product):
        if j == product_x:
            recommendation_list.append(list(sorted_rules.iloc[i]["consequents"])[0])
```

## Functionalization of the Work

```python
def recommender_products(product_x, th=0.001):
  df = pd.read_csv("/kaggle/input/groceries-dataset/Groceries_dataset.csv")
  df['Date'] = pd.to_datetime(df['Date'])
  df["invoice"] = df['Member_number'].astype(str) + '_' + df["Date"].astype(str)
  #Pivot Table
  new_df = pd.pivot_table(df, index='invoice', columns='itemDescription', values= 'Member_number')
  new_df = new_df +1
  new_df = new_df.fillna(0).applymap(lambda x: 1 if x!= 0 else 0)
  #Association Rules
  apriori_df = apriori(new_df,
                     min_support= th,
                     use_colnames=True,
                     low_memory=True)
  rules = association_rules(apriori_df,
                          metric='support',
                          min_threshold= th)
  sorted_rules = rules.sort_values('lift', ascending=False)
  #Recommendation List
  recommendation_list = []
  for i, product in enumerate(sorted_rules["antecedents"]):
      for j in list(product):
          if j == product_x:
             recommendation_list.append(list(sorted_rules.iloc[i]["consequents"])[0])
  final_recommendation= []
  for i in recommendation_list:
      if i not in final_recommendation:
        final_recommendation.append(i)
  return final_recommendation[0:3]
```

```python
recommender_products('sausage')
```
